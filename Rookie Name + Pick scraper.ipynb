```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/2024_NBA_draft'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify)
```

```python
table = soup.find_all('table')[3]
print(table)
```

```python
td_tags = table.find_all('td')[2::7]
print(td_tags)
```

```python
td_tags.extend(table.find_all('td')[339:400:7])
td_tags.extend(table.find_all('td')[403:410:7])
print(td_tags)
#I could probably clean this up and not use the extend and just change the index on the td_tags variable but I want to
#keep this as an example for myself
```

```python
a_tags = []
for td in td_tags:
    a_tags.extend(td.find_all('a'))
```

```python
rookies = [name.text.strip() for name in a_tags]
print(rookies)
```

```python
rookies_string = ', '.join(rookies)
print(rookies_string)
```

```python
rookies_list = [rookie.strip() for rookie in rookies_string.split(',')]
```

```python
df = pd.DataFrame({
    'Name': rookies_list,
    'Draft Pick': range(1,59)
    })
df.to_csv(r"D:\Personal_Python_Projects\venv\Projects\NBA_ROTY_Prediction\Rookie_Names.csv", index=False)
df
```
