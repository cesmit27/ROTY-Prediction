```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.basketball-reference.com/awards/roy.html'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
table = soup.find_all('table')[0]
print(table)
```

```python
a_tags = table.find_all('a')
print(a_tags)
```

```python
players = [name.text.strip() for name in a_tags][2:150:5]
print(players)
```

```python
players_list = [player.strip() for player in players_string.split(',')]
print(players_list)
```
